# 1. Write a Python Program to find sum of array?

```python
def sumofarray():
    in_arry= eval(input("Enter the array: "))
    print(f"The sum of {in_arry} is {sum(in_arry)}")
sumofarray()
```

# 2. Write a Python Program to find largest element in an array?

```python
def largestElement():
    in_arry=eval(input("Enter the array: "))
    print(f"The largest element in {in_arry} is {max(in_arry)}")

largestElement()
```

# 3. Write a Python Program for array rotation?

```python
def reverseofarray():
    in_arry=eval(input("Enter the array: "))
    print(f"The reverse of array {in_arry} is {in_arry[::-1]}")
reverseofarray()
```

# 4. Write a Python Program to Split the array and add the first part to the end?

# 5. Write a Python Program to check if given array is Monotonic?

```python
def checkmonotonic():
    in_arry= eval(input("Enter the array: "))
    if(all(in_arry[i]<=in_arry[i+1] 
           for i in range(len(in_arry)-1)) 
         or
         all(in_arry[i]>=in_arry[i+1] 
             for i in range(len(in_arry)-1))):
            print(f'Array {in_arry} is Monotonic')
    else:
        print(f'Array {in_arry} is Not Monotonic')
checkmonotonic()
```
