# 1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
def fibo(n):
    if n <= 1:
        return n 
    else:
        return fibo(n-1) + fibo(n-2)
sequence = int(input("enter a number to find the fibonacci series : "))

if sequence <=0:
    print("Please enter a positive number")
elif sequence == 1:
    print("fibonacci series upto ",sequence, ":")
    print(1)
else:
    print("Fibonacci series upto ", sequence, ":")
    for i in range(sequence):
        print(fibo(i),end=" ")
```

# 2. Write a Python Program to Find Factorial of Number Using Recursion?

```python
def fact(n):
    if n == 1:
        return n
    else:
        return n * fact(n-1)
    
num = int(input("Enter a Number to find factorial : "))
if num < 0:
    print("Can not calculate Factorial of a negactive number")
elif num == 0:
    print("Factorial of zero is 1")
else:
    print("Factorial of {} is {}".format(num,fact(num)))
```

# 5. Write a Python Program for cube sum of first n natural numbers?

```python
n = int(input("enter number : "))
sum = 0
for i in range (1, n+1):
    sum+=i**3
print("Cube sum of {} natural number is  {}".format(n,sum))
```
