# 1. Write a Python program to convert kilometers to miles?

```python
kilometers = float(input("Enter Kilometer :"))
conversion_unit= 0.621
miles = kilometers * conversion_unit
print("%0.3f kilomters is equal to %0.3f miles"%(kilometers,miles))
```

# 2. Write a Python program to convert Celsius to Fahrenheit?

```python
celsius = float(input("Celsius :"))
fahrenheit = (celsius * 1.8) + 32
print("%0.2f celsius is %0.2f fahrenheit"%(celsius,fahrenheit))
```

# 4. Write a Python program to solve quadratic equation?

- quadratic equation ax^2+bx+c=0
- a, b, c =   known numbers, where a â‰  0
- x   =   the unknown
- descriminant = b**2 - (4* a * c)
- x1 = (-b-sqrt(d))/2 , x2 = (b-sqrt(d))/2

```python
import cmath

a =int(input("Enter value of a :"))
b =int(input("Enter value of b :"))
c =int(input("Enter value of c :"))

dis = (b**2) - (4 * a*c)

x1 = (-b - cmath.sqrt(dis))/(2 * a)
x2 = (-b + cmath.sqrt(dis))/(2 * a)

print("The root are")
print(x1)
print(x2)
```

# 5. Write a Python program to swap two variables without temp variable?

```python
x = int(input("Num1 : "))
y = int(input("Num2 : "))
print("Before swaping x = {}, y = {}".format(x,y))
x,y=y,x
print("After swaping x = {}, y = {}".format(x,y))
```
